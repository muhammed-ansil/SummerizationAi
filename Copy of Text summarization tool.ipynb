```python
import nltk
from sumy.parsers.plaintext import PlaintextParser
from sumy.parsers.html import HtmlParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

# Ensure NLTK resources are downloaded
nltk.download('punkt')

# Function to summarize text from a URL
import requests
from sumy.parsers.html import HtmlParser

def summarize_from_url(url, sentences_count=3):
    try:
        # Parse the HTML content from the URL
        parser = HtmlParser.from_url(url, Tokenizer("english"))

        # Initialize the summarizer
        summarizer = LexRankSummarizer()

        # Generate the summary
        summary = summarizer(parser.document, sentences_count=sentences_count)

        # Print the summary
        print("\nSummary:")
        for sentence in summary:
            print(sentence)
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e}")
    except requests.exceptions.ConnectionError as e:
        print(f"Connection Error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")


# Function to summarize text from plain string
def summarize_from_text(text, sentences_count=3):
    # Parse the text
    parser = PlaintextParser.from_string(text, Tokenizer("english"))

    # Initialize the summarizer
    summarizer = LexRankSummarizer()

    # Generate the summary
    summary = summarizer(parser.document, sentences_count=sentences_count)

    # Print the summary
    print("\nSummary:")
    for sentence in summary:
        print(sentence)

# Example Usage:

# Summarize from a plain text (replace with your actual article)
text = """
This is a sample article. It contains multiple sentences that will be used to generate a summary.
The purpose of this script is to demonstrate how to use the sumy library for text summarization.
Sumy provides several summarization algorithms, including LexRank, which we will be using here.
LexRank is a graph-based algorithm that ranks sentences based on their importance in the text.
This script takes the input text, tokenizes it, and then uses the LexRank summarizer to generate a summary.
The generated summary will contain the most important sentences from the input text.
The length of the summary can be controlled by the user, and we can try different algorithms and see the results.
This is another sentence that provides additional context for the article.
It will be interesting to see how the algorithm chooses the important sentences for the summary.
"""
summarize_from_text(text, sentences_count=3)

# Summarize from a URL (replace with your actual URL)
url = "https://www.ibm.com/think/topics/data-science"  # Replace with a valid article URL
summarize_from_url(url, sentences_count=3)
```
